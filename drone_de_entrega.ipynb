## Ejercicio del drone de entrega

- Considera un drone que hace entregas, se mueve en un grid de $N \times N$ .
- Para cada paso de tiempo (time step), el drone se mueve a un cuadrado adyacente (arriba, abajo, izquierda, derecha).
- Todas las acciones cuestan 1. Se tienen k destinos fijos en el grid.
- El drone debe visitarlos todos al menos una vez.
- El drone comienza con una carga de bateria de C unidades y para que pueda realizar cada paso de tiempo debe tener al menos 1 unidad de carga.
- El drone pierde 1 unidad de carga por cada acción.
- Existen p celdas de carga (power cells) en el grid.
- Cuando el drone visita una de esas celdas, la carga de su batería se resetea a C unidades y esa celda queda inutilizable.

### Respuestas a las preguntas

**a) ¿Cuál es el máximo factor de ramificación?**
    Es 4. En cada paso se mueve a un cuadrado adyacente (arriba, abajo, izquierda, derecha). Esto significa que no puede quedarse quieto, ni teletransportarse ni usar batería como acción separada. Entonces, en un grid $ N \times N$ hay tres posibilidades. Si está en una esquina, tiene dos opciones de movimiento. Si está en un borde que no es esquina,  puede hacer tres movimientos distintos. Y si está en el interior del grid, a lo más, puede hacer cuatro movimientos. Por eso el máximo factor de ramificación es 4.

**b) ¿Cuál es el tamaño del espacio de estado más pequeño de este problema?**
    Un estado tiene que contener toda la información necesaria para decidir el futuro. Primero necesitamos la posición en el grid: $(N^2)$. Luego, para el conjunto de destinos ya visitados, consideramos que hay un total de k destinos, cada uno puede estar visitado o no visitado, así hay $(2^k)$ combinaciones. Después, hay $p$ celdas de carga y cada una puede estar o disponible o utilizada, entonces hay un total de $(2^p)$ power cells aún utilizables. Por último, el nivel de batería puede tomar valores $0,..., C$ ($C+1$ valores). Así, obtenemos el tamaño mínimo del espacio de estados:
    $$|S| = N^2 · (C + 1) · 2^{k+p}$$ 
    (Pero, si sólo contáramos estados desde los que se puede actuar (batería $≥ 1, C$ valores):
    $|S| = N^2 · C · 2^{k+p}$.)
